```python
#Make FRP as a list
frp_list = [];
for i in range(len(firmsx)):
  key = firmsx.iloc[i]['FRP']
  key = round(key)
  if( 1.0 <= key <= 5.0 ):
    frp_list.append(1)
  elif ( 6.0 <= key <= 10.0 ):
    frp_list.append(2)
  elif ( 11.0 <= key <= 20.0 ):
    frp_list.append(3)
  elif ( 21.0 <= key <= 30.0 ):
    frp_list.append(4)
  elif ( 31.0 <= key <= 40.0 ):
    frp_list.append(5)
  elif ( 41.0 <= key <= 50.0 ):
    frp_list.append(6)
  elif ( 51.0 <= key <= 60.0 ):
    frp_list.append(7)
  elif ( 61.0 <= key <= 70.0 ):
    frp_list.append(8)
  elif ( 71.0 <= key <= 80.0 ):
    frp_list.append(9)
  elif ( 81.0 <= key <= 90.0 ):
    frp_list.append(10)
  elif ( 91.0 <= key <= 100.0 ):
    frp_list.append(11)
  elif ( key >= 101.0 ):
    frp_list.append(12)
```

```python
len(frp_list)
```

```python
firmsx['FRP'] = frp_list
firmsx['FRP'].astype('int64')
```

```python
# import module
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent="geoapiExercises")

#Make STATE  list
state_list = [];
country_list = [];

for i in range(len(firmsx)):
  Latitude = firmsx.iloc[i]['LATITUDE']
  Longitude = firmsx.iloc[i]['LONGITUDE']
  location = geolocator.reverse(str(Latitude)+","+str(Longitude))
  if(location is not None):
     address = location.raw['address']
     state = address.get('state', '')
     country = address.get('country', '')
     print(Latitude, Longitude, state, country)
     state_list.append(state)
     country_list.append(country)
  else:
    print(Latitude, Longitude, "No Result", "No Result")
    state_list.append("No Result")
    country_list.append("No Result")
```

```python
len(state_list)
```

```python
len(country_list)
```

```python
firmsx['STATE'] = state_list
firmsx['COUNTRY'] = country_list
firmsx
```
